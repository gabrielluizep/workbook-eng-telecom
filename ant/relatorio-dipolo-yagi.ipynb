```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({"axes.grid": True, "font.size": 14})
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 6))

axs[0].plot(f, g_dipole)
axs[0].set_title("Antena Dipolo")
axs[0].set_ylabel("Ganho (dBm)")
axs[0].set_xlabel("Frequncia (MHz)")

axs[1].plot(f, g_yagi, label="Yagi")
axs[1].set_title("Antena Yagi")
axs[1].set_ylabel("Ganho (dBm)")
axs[1].set_xlabel("Frequncia (MHz)")

fig.tight_layout(pad=3)
plt.show()
```

```python
angles = [
    0,
    20,
    40,
    60,
    80,
    100,
    120,
    140,
    160,
    180,
    200,
    220,
    240,
    260,
    280,
    300,
    320,
    340,
    0,
]

p_dipole = [
    -45,
    -44,
    -43,
    -44,
    -50,
    -44,
    -44,
    -45,
    -50,
    -48,
    -54,
    -49,
    -45,
    -41,
    -49,
    -42,
    -41,
    -46,
    -45,
]

p_yagi = [
    -51,
    -50,
    -50,
    -60,
    -50,
    -48,
    -48,
    -46,
    -46,
    -47,
    -55,
    -47,
    -41,
    -40,
    -41,
    -44,
    -51,
    -59,
    -51,
]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={"polar": True})

# tunr angles to radians
radians = np.array(angles) * np.pi / 180

# normalize power, so the maximum is 0
p_dipole = np.array(p_dipole) - max(p_dipole)
p_yagi = np.array(p_yagi) - max(p_yagi)

axs[0].plot(radians, p_dipole)
axs[0].set_title("Antena Dipolo")
axs[0].set_rlabel_position(-30)

axs[1].plot(radians, p_yagi, label="Yagi")
axs[1].set_title("Antena Yagi")
axs[1].set_rlabel_position(-30)


fig.tight_layout(pad=3)
plt.show()
```
