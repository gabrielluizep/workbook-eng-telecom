```python
import numpy as np
import matplotlib.pyplot as plt

P_t = 43.01
PL_d_0 = 120
d_0 = 1
n = 3.5
P_n = -120

ds = np.arange(0.1, 11, 0.1)
snrs = np.zeros(len(ds))
for i, d in enumerate(ds):
    PL = PL_d_0 + 10 * n * np.log10(d / d_0)
    P_r = P_t - PL
    snrs[i] = P_r - P_n

snrs[snrs > 50] = 50
snrs[snrs < 0] = 0

plt.figure()
plt.plot(ds, snrs)
plt.ylabel("Taxa [Mbps]")
plt.xlabel("Distância da ERB [km]")
plt.show()
```
