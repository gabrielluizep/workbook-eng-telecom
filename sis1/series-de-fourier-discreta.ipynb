```python
from numpy import pi, e
import matplotlib.pyplot as plt

N = 7
w_0 = 2 * pi / N


def a_k(k):
    r = e ** (-1j * k * w_0)
    return (1 / 7) * (1 - r**5) / (1 - r)


def x_n(n):
    return sum([a_k(k) * e ** (1j * k * w_0 * n) for k in range(7, 14)])


n = range(-10, 10)

plt.stem(n, [x_n(n) for n in n])
plt.xticks(range(-10, 10))
plt.show()
plt.savefig('a.png')
```

```python
from numpy import pi, e
import matplotlib.pyplot as plt

N = 6
w_0 = 2 * pi / N


def a_k(k):
    r = e ** (-1j * k * w_0)
    return (1 / 6) * (1 + 2 * r - r**2 - r**4 + 2 * r**5)


def x_n(n):
    return sum([a_k(k) * e ** (1j * k * w_0 * n) for k in range(0, 6)])


n = range(-10, 10)

plt.stem(n, [x_n(n) for n in n])
plt.xticks(range(-10, 10))
plt.show()
plt.savefig('c.png')
```
