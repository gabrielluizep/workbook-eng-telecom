### PCM30

| Timeslots              | 32           |
|------------------------|---------     |
| Time slot 0            | Alinhamento  |
| Time slot 16           | Sinalização  |
| Timeslot 1-15 e 17-31  | Voz          |
| Bits por time slot     | 8 bits       |
| Tamanho do quadro      | 256 bits     |
| Palavra de alinhamento | x0011011     |

```python
def order_ts(dado, paq):
    # first and second frame after positions
    first = 256
    second = 512

    ts = []
    if dado[first : first + 1] == "1" and dado[second : second + 8] == paq:
        ts = [dado[j : j + 8] for j in range(0, second, 8)]

    return ts


df_data = []
with open("./data/RX(vetor)MQ_v2.txt") as arq:
    data = arq.read()
    data = data.replace(" ", "")
    start = 0

    PAQ = "10011011"
    frame = 0

    while True:
        start = data.find(PAQ, start)
        if start == -1:
            break

        ts = order_ts(data[start:], PAQ)
        if len(ts) > 0:
            PAQ = ts.pop(0)
            frame += 1
            frame_data = [PAQ]
            for i, byte in enumerate(ts):
                if i > 30:
                    if i == 31:
                        frame += 1
                    ts = i - 32
                else:
                    ts = i
                frame_data.append(byte)
            df_data.append(frame_data)
        start += 1
```

```python
temp = []

for frame in df_data:
    temp.append(frame[:32])
    temp.append(frame[32:])


df = pd.DataFrame(temp)
df.index.name = "Frame"
df.columns.name = "Timeslot"

df.to_csv("./data/RX(vetor)MQ_v2.csv")

df
```
