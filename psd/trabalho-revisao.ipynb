```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["axes.grid"] = True
```

```python
n = np.arange(-5, 6)
x = lambda n: (2 - n) * (np.heaviside(n, 1) - np.heaviside(n - 4, 1))
y = x(-2 * n - 4)

fig, ax = plt.subplots(1, 2, figsize=(18, 6))

ax[0].stem(n, x(n))
ax[0].set_title(r"$x[n] = (2 - n) \cdot (u[n] - u[n-4])$")
ax[0].set_xlabel(r"$n$")
ax[0].set_ylabel(r"$x[n]$")


ax[1].stem(n, y)
ax[1].set_title(r"$y[n] = x[-2n - 4]$")
ax[1].set_xlabel(r"$n$")
ax[1].set_ylabel(r"$y[n]$")

plt.show()
```
