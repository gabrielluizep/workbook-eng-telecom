### Exemplo 

Usando a definição calcular a DFT de $x[n] = 4\delta[n] + 3\delta[n-1] + 2\delta[n-2] + \delta[n-3]$  

```python
x_n = [4, 3, 2, 1]
N = len(x_n)

X_k = np.zeros(N, dtype=complex)
for k in range(N):
    for n in range(N):
        X_k[k] += x_n[n] * np.exp(-2j * np.pi * k * n / N)

print(X_k)
```

```python
X_k = np.fft.fft(x_n)

print(X_k)
```

### Exemplo 02

Calcular a DFT de $x[n] = \delta[n]$

### Exemplo 06

Considere a sequência $x[n] = 4\delta[n] + 3\delta[n-1] +2\delta[n-2] + \delta[n-3]$. Seja $X[k]$ a DFT de 6 pontos de $x[n]$. Encontra a sequência finita $y[n]$ cuja DFT de 6 pontos é $Y[k] = e^{j\frac{2\pi}{6}k}X[k]$.

```python
N = 6
n = np.arange(N)
k = np.arange(N)
x_n = [4, 3, 2, 1, 0, 0]

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(18, 4))
fig.tight_layout(h_pad=4, w_pad=3)

axs[0, 0].stem(n, x_n)
axs[0, 0].set_title("Input")
axs[0, 0].set_xlabel(r"$n$")
axs[0, 0].set_ylabel(r"$x[n]$")

X_k = np.fft.fft(x_n)

axs[0, 1].stem(k, np.abs(X_k))
axs[0, 1].set_title("Magnitude")
axs[0, 1].set_xlabel(r"$k$")
axs[0, 1].set_ylabel(r"$|X[k]|$")

axs[0, 2].stem(k, np.angle(X_k))
axs[0, 2].set_title("Phase")
axs[0, 2].set_xlabel(r"$k$")
axs[0, 2].set_ylabel(r"$∠X[k]$")

Y_k_fft = np.exp(2j * np.pi / 6 * 4 * k) * X_k
y_n_fft = np.fft.ifft(Y_k_fft)

axs[1, 0].stem(n, y_n_fft)
axs[1, 0].set_title("Output")
axs[1, 0].set_xlabel(r"$n$")
axs[1, 0].set_ylabel(r"$y[n]$")

axs[1, 1].stem(k, np.abs(Y_k_fft))
axs[1, 1].set_title("Magnitude")
axs[1, 1].set_xlabel(r"$k$")
axs[1, 1].set_ylabel(r"$|Y[k]|$")

axs[1, 2].stem(k, np.angle(Y_k_fft))
axs[1, 2].set_title("Phase")
axs[1, 2].set_xlabel(r"$k$")
axs[1, 2].set_ylabel(r"$∠Y[k]$")

plt.show()
```

### Exemplo 07

Seja a sequência $x[n] = 2\delta[n]+\delta[n-1] + \delta[n-3]$. A DFT de 5 pontos de $x[n]$ é computada e a sequência resultante é elevada ao quadrado: $Y[k] = X[k]^2$. A seguir a DFT inversa é calculada, resultando a sequência $y[n]$. Encontre a sequência $y[n]$.

```python
N = 5
n = np.arange(N)
x_n = [2, 1, 0, 1, 0]

fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(18, 8))
fig.tight_layout(h_pad=4, w_pad=3)

axs[0, 0].stem(n, x_n)
axs[0, 0].set_title("Input")
axs[0, 0].set_xlabel(r"$n$")
axs[0, 0].set_ylabel(r"$x[n]$")

X_k = np.fft.fft(x_n)

axs[0, 1].stem(n, np.abs(X_k))
axs[0, 1].set_title("Magnitude")
axs[0, 1].set_xlabel(r"$k$")
axs[0, 1].set_ylabel(r"$|X[k]|$")

axs[0, 2].stem(n, np.angle(X_k))
axs[0, 2].set_title("Phase")
axs[0, 2].set_xlabel(r"$k$")

Y_k_fft = X_k**2
y_n_fft = np.fft.ifft(Y_k_fft)

axs[1, 0].stem(n, y_n_fft)
axs[1, 0].set_title("Output FFT")
axs[1, 0].set_xlabel(r"$n$")
axs[1, 0].set_ylabel(r"$y[n]$")

axs[1, 1].stem(n, np.abs(Y_k_fft))
axs[1, 1].set_title("Magnitude")
axs[1, 1].set_xlabel(r"$k$")
axs[1, 1].set_ylabel(r"$|Y[k]|$")

axs[1, 2].stem(n, np.angle(Y_k_fft))
axs[1, 2].set_title("Phase")
axs[1, 2].set_xlabel(r"$k$")
axs[1, 2].set_ylabel(r"$∠Y[k]$")

y_n = ndimage.convolve(x_n, x_n, mode="wrap", origin=-int(N / 2))
Y_k = np.fft.fft(y_n)

axs[2, 0].stem(n, y_n)
axs[2, 0].set_title("Output Convolution")
axs[2, 0].set_xlabel(r"$n$")
axs[2, 0].set_ylabel(r"$y[n]$")

axs[2, 1].stem(n, np.abs(Y_k))
axs[2, 1].set_title("Magnitude")
axs[2, 1].set_xlabel(r"$k$")
axs[2, 1].set_ylabel(r"$|Y[k]|$")

axs[2, 2].stem(n, np.angle(Y_k))
axs[2, 2].set_title("Phase")
axs[2, 2].set_xlabel(r"$k$")
axs[2, 2].set_ylabel(r"$∠Y[k]$")


plt.show()
```

### Exemplo 08

Considere as duas sequências

$$
x[n] = \delta[n] + 3\delta[n-1] + 3\delta[n-2] + 2\delta[n-3] \\

\\ \text{e} \\

h[n] = \delta[n] + \delta[n-1] + \delta[n-2] + \delta[n-3]
$$

Se formarmos o produto $Y[k]=X[k]H[k]$, onde $X[k]$ e $H[k]$ são as DFTs de 5 pontos de $x[n]$ e $h[n]$, respectivamente, e tomarmos a DFT inversa, teremos $y[n]$. Encontre a sequência $y[n]$.


```python
N = 5
n = np.arange(N)
x_n = [1, 3, 3, 2, 0]
h_n = 4 * [1] + [0]

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(18, 4))

axs[0].stem(n, x_n)
axs[0].set_title("Input")
axs[0].set_xlabel(r"$n$")
axs[0].set_ylabel(r"$x[n]$")

axs[1].stem(n, h_n)
axs[1].set_title("Impulse response")
axs[1].set_xlabel(r"$n$")
axs[1].set_ylabel(r"$h[n]$")

X_k = np.fft.fft(x_n)
H_k = np.fft.fft(h_n)
Y_k = X_k * H_k
y_n = np.fft.ifft(Y_k)

axs[2].stem(n, y_n)
axs[2].set_title("Output FFT")
axs[2].set_xlabel(r"$n$")
axs[2].set_ylabel(r"$y[n]$")

y_n_conv = ndimage.convolve(x_n, h_n, mode="wrap", origin=-int(N / 2))

axs[3].stem(n, y_n_conv)
axs[3].set_title("Output Convolution")
axs[3].set_xlabel(r"$n$")
axs[3].set_ylabel(r"$y[n]$")

plt.show()
```

### Exemplo 09

Encontre a DFT inversa de $Y[k] = |X[k]|^2$, onde $X[k]$ é a DFT de 10 pontos da sequência $x[n] = u[n] - u[n-6]$.

```python
N = 10
n = np.arange(N)

x_n = np.heaviside(n, 1) - np.heaviside(n - 6, 1)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(18, 4))

axs[0].stem(n, x_n)
axs[0].set_title("Input")
axs[0].set_xlabel(r"$n$")
axs[0].set_ylabel(r"$x[n]$")

X_k = np.fft.fft(x_n)
Y_k = X_k * np.conj(X_k)

y_n = np.fft.ifft(Y_k)

axs[1].stem(n, y_n)
axs[1].set_title("Output")
axs[1].set_xlabel(r"$n$")
axs[1].set_ylabel(r"$y[n]$")

plt.show()
```

### Exemplo 09

Suponha que o sinal $x(t) = sin(2*\pi*5*t) + 2 sin(2*\pi*50*t)$, amostrado a $F_s = 1000\text{Hz}$ é corrompido por uma pequena quantidade de ruído. Esse sinal é gerado pelos seguintes comandos:

```python
A1 = 5
f1 = 5

A2 = 2
f2 = 50

fs = 1e3

t = np.arange(0, 1, 1 / fs)
f = np.arange(-fs / 2, fs / 2, 1)

x_1 = A1 * np.sin(2 * np.pi * f1 * t)
x_2 = A2 * np.sin(2 * np.pi * f2 * t)

noise = np.random.normal(0, 1, len(t))

x_clean = x_1 + x_2
x_noisy = x_clean + noise

fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(18, 12))
fig.tight_layout(h_pad=4, w_pad=3)

axs[0, 0].plot(t, x_clean)
axs[0, 0].set_title("Clean signal")
axs[0, 0].set_xlabel(r"$t$")
axs[0, 0].set_ylabel(r"$x(t)$")

X_clean = np.fft.fftshift(np.fft.fft(x_clean))

axs[0, 1].plot(f, np.abs(X_clean))
axs[0, 1].set_title("Clean signal FFT")
axs[0, 1].set_xlabel(r"$k$")
axs[0, 1].set_ylabel(r"$|X[k]|$")
axs[0, 1].set_xlim([-100, 100])

axs[1, 0].plot(t, x_noisy)
axs[1, 0].set_title("Noisy signal")
axs[1, 0].set_xlabel(r"$t$")
axs[1, 0].set_ylabel(r"$x(t)$")

X_noisy = np.fft.fftshift(np.fft.fft(x_noisy))

axs[1, 1].plot(f, np.abs(X_noisy))
axs[1, 1].set_title("Noisy signal FFT")
axs[1, 1].set_xlabel(r"$k$")
axs[1, 1].set_ylabel(r"$|X[k]|$")
axs[1, 1].set_xlim([-100, 100])

N = 10
b = np.ones(N)
y_n = signal.lfilter(b, 1, x_noisy)
Y_k = np.fft.fftshift(np.fft.fft(y_n))

axs[2, 0].plot(t, y_n)
axs[2, 0].set_title("Filtered signal")
axs[2, 0].set_xlabel(r"$t$")
axs[2, 0].set_ylabel(r"$y(t)$")

axs[2, 1].plot(f, np.abs(Y_k))
axs[2, 1].set_title("Filtered signal FFT")
axs[2, 1].set_xlabel(r"$k$")
axs[2, 1].set_ylabel(r"$|Y[k]|$")
axs[2, 1].set_xlim([-100, 100])


plt.show()
```
