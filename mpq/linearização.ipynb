```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

# plt.style.use(["ieee"])
# plt.rcParams["font.family"] = "DejaVu Serif"
```

# Exemplos

```python
df = pd.DataFrame(
    {
        "Aresta (cm)": [2.00, 4.00, 6.00, 8.00],
        "Massa (g)": [0.030, 0.123, 0.286, 0.491],
    }
)

df["x"] = df["Aresta (cm)"]
df["y"] = df["Massa (g)"]
df["x'"] = df["x"] ** 2

n = df.shape[0]
sx = df["x'"].sum()
sy = df["y"].sum()
sxsq = (df["x'"] ** 2).sum()
sxy = (df["x'"] * df["y"]).sum()

A = (n * sxy - sx * sy) / (n * sxsq - sx**2)
B = (sy * sxsq - sx * sxy) / (n * sxsq - sx**2)

print(f"a = {A}, b = {B}")

df["y'"] = A * df["x'"] + B

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(df["x"], df["y"], "b", label="Dados Originais")

axs[0].legend()
axs[0].grid()

axs[1].plot(df["x'"], df["y"], "b", label="Dados Transformados")
axs[1].plot(df["x'"], df["y'"], "r", label="MMQ")
axs[1].legend()
axs[1].grid()
```

```python
# Tempo (s) 0 60 120 180 240 300 360 420 480 540 600
# Temperatura (C) 89,2 85,0 82,2 80,1 77,9 76,2 74,1 72,3 71,0 69,6 68,0

df = pd.DataFrame(
    {
        "Tempo (s)": [0, 60, 120, 180, 240, 300, 360, 420, 480, 540, 600],
        "Temperatura (C)": [
            89.2,
            85.0,
            82.2,
            80.1,
            77.9,
            76.2,
            74.1,
            72.3,
            71.0,
            69.6,
            68.0,
        ],
    }
)

df["x"] = df["Tempo (s)"]
df["y"] = df["Temperatura (C)"]
df["x'"] = df["x"]
df["y'"] = np.log(df["y"])

n = df.shape[0]
sx = df["x'"].sum()
sy = df["y"].sum()
sxsq = (df["x'"] ** 2).sum()
sxy = (df["x'"] * df["y"]).sum()

A = (n * sxy - sx * sy) / (n * sxsq - sx**2)
B = (sy * sxsq - sx * sxy) / (n * sxsq - sx**2)

print(f"a = {A}, b = {B}")

df["y''"] = A * df["x'"] + B

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(df["x"], df["y"], "b", label="Dados Originais")
axs[0].set_ylabel("Temperatura (C)")
axs[0].set_xlabel("Tempo (s)")
axs[0].legend()
axs[0].grid()

axs[1].plot(df["x"], df["y'"], "b", label="Dados Transformados")
# axs[1].plot(df["x'"], df["y''"], "r", label="MMQ")
axs[1].set_ylabel("Temperatura (C)")
axs[1].set_xlabel("Tempo (s)")
axs[1].legend()
axs[1].grid()
```

1) Num experimento sobre MRUV, um grupo de alunos obteve as seguintes medidas:

x (cm) 8,0 61,0 200,0 317,0 402,0

t (s) 2,0 8,0 15,0 18,0 20,0

a) Linearize a equao e indique os coeficientes angular e linear, bem como as variveis dependente e independente;

b) Atravs do mtodo dos mnimos quadrados, determine a equao da melhor reta e a represente graficamente, junto com os pontos experimentais.

```python
df = pd.DataFrame(
    {
        "x (cm)": [8.0, 61.0, 200.0, 317.0, 402.0],
        "t (s)": [2.0, 8.0, 15.0, 18.0, 20.0],
    }
)

df["y"] = df["x (cm)"]
df["x"] = df["t (s)"]
df["x'"] = df["x"] ** 2

n = df.shape[0]
sx = df["x'"].sum()
sy = df["y"].sum()
sxsq = (df["x'"] ** 2).sum()
sxy = (df["x'"] * df["y"]).sum()

A = (n * sxy - sx * sy) / (n * sxsq - sx**2)
B = (sy * sxsq - sx * sxy) / (n * sxsq - sx**2)

print(f"a = {A}, b = {B}")

df["y'"] = A * df["x'"] + B
df["y_ajustado"] = np.sqrt(df["y'"])

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(df["x"], df["y"], "bo", label="Dados Originais")
axs[0].plot(df["x"], df["y'"], "ro")
axs[0].set_ylabel("Tempo (s)")
axs[0].set_xlabel("Distncia (cm)")
axs[0].legend()
axs[0].grid()

axs[1].plot(df["x'"], df["y"], "bo", label="Dados Linealizados")
axs[1].plot(df["x'"], df["y'"], "ro", label="MMQ")
axs[1].set_ylabel("Tempo (s^2)")
axs[1].set_xlabel("Distncia (cm)")
axs[1].legend()
axs[1].grid()
```

2) Os dados abaixo tabelados esto relacionados por uma equao do tipo: y(x) = ax^n

y (L) 3,21 5,31 8,23 15,00 26,10 53,80

x (h) 1,69 4,93 10,97 28,47 88,83 288,00

a) Linearize a equao e indique os coeficientes angular e linear, bem como as variveis dependente e independente;

b) Atravs do mtodo dos mnimos quadrados, determine a equao da melhor reta e a represente graficamente, junto com os pontos experimentais.


```python
df = pd.DataFrame(
    {
        "y (L)": [3.21, 5.31, 8.23, 15.00, 26.10, 53.80],
        "x (h)": [1.69, 4.93, 10.97, 28.47, 88.83, 288.00],
    }
)

df["x"] = df["y (L)"]
df["y"] = df["x (h)"]
df["x_lin"] = np.log(df["x"])
df["y_lin"] = np.log(df["y"])

n = df.shape[0]
sx = df["x_lin"].sum()
sy = df["y_lin"].sum()
sxsq = (df["x_lin"] ** 2).sum()
sxy = (df["x_lin"] * df["y_lin"]).sum()

A = (n * sxy - sx * sy) / (n * sxsq - sx**2)
B = (sy * sxsq - sx * sxy) / (n * sxsq - sx**2)

df["y_lin_est"] = A * df["x_lin"] + B
df["y_est"] = np.exp(df["y_lin_est"])

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(df["x"], df["y"], "bo", label="Dados Originais")
axs[0].plot(df["x"], df["y_est"], "ro")
axs[0].grid()

axs[1].plot(df["x_lin"], df["y_lin"], "bo", label="Dados Transformados")
axs[1].plot(df["x_lin"], df["y_lin_est"], "ro")

axs[1].grid()
```

3) Os dados abaixo tabelados esto relacionados por uma equao do tipo: y(x) = ae^(bx).

y (mC) 2410 826 419 348 104 22

x (s) 1,37 3,39 4,57 4,71 7,02 9,48

a) Linearize a equao e indique os coeficientes angular e linear, bem como as variveis dependente e independente;

b) Atravs do mtodo dos mnimos quadrados (M.M.Q.), determine a equao da melhor reta e a represente graficamente, junto com os pontos experimentais.

```python
# y (mC) 2410 826 419 348 104 22
# x (s) 1,37 3,39 4,57 4,71 7,02 9,48

df = pd.DataFrame(
    {
        "y (mC)": [2410, 826, 419, 348, 104, 22],
        "x (s)": [1.37, 3.39, 4.57, 4.71, 7.02, 9.48],
    }
)

df["y"] = df["y (mC)"]
df["x"] = df["x (s)"]
df["x_lin"] = df["x"]
df["y_lin"] = np.log(df["y"])

n = df.shape[0]
sx = df["x_lin"].sum()
sy = df["y_lin"].sum()
sxsq = (df["x_lin"] ** 2).sum()
sxy = (df["x_lin"] * df["y_lin"]).sum()

A = (n * sxy - sx * sy) / (n * sxsq - sx**2)
B = (sy * sxsq - sx * sxy) / (n * sxsq - sx**2)

df["y_lin_est"] = A * df["x_lin"] + B
df["y_est"] = np.exp(df["y_lin_est"])

print(f"a = {A}, b = {B}")

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(df["x"], df["y"], "bo", label="Dados Originais")
axs[0].plot(df["x"], df["y_est"], "ro")
axs[0].grid()

axs[1].plot(df["x_lin"], df["y_lin"], "b", label="Dados Transformados")
axs[1].plot(df["x_lin"], df["y_lin_est"], "r")

axs[1].grid()
```
