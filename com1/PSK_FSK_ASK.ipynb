```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi, cos

plt.rcParams.update({"axes.grid": True, "font.size": 14})
```

### PSK

```python
fs = 2500  # Sampling frequency 2500 Hz
Ts = 1 / fs  # Sampling interval 0.4 ms

t = np.arange(0, 1, Ts)  # Time vector

fc = 50  # Carrier frequency 50 Hz
Tc = 1 / fc  # Carrier period
Ac = 1  # Carrier amplitude

# Binary signal generation
Nb = 10  # Number of bits
Nsym = 2  # Number of bits per symbol

data = np.random.randint(0, 2, Nb)  # Random binary signal
data = np.repeat(data, fs / Nb)  # Repetition of the binary signal
```

```python
phi_t = 2 * pi * data / Nsym  # Phase shift
psk = Ac * cos(2 * pi * fc * t + phi_t)  # PSK signal

w_t = 2 * pi * (data + 1) * fc * t  # Phase shift
fsk = Ac * cos(w_t)  # FSK signal

ask = Ac * data * cos(2 * pi * fc * t)  # ASK signal
```

```python
fig, ax = plt.subplots(4, 1, figsize=(20, 12))
fig.tight_layout(pad=2.0)


ax[0].set_title("Binary signal")
ax[0].set_xlabel("$t$")
ax[0].set_ylabel("$d_i(t)$")
ax[0].plot(t, data)

ax[1].set_xlabel("$t$")
ax[1].set_ylabel("$PSK$")
ax[1].plot(t, psk)

ax[2].set_xlabel("$t$")
ax[2].set_ylabel("$FSK$")
ax[2].plot(t, fsk)

ax[3].set_xlabel("$t$")
ax[3].set_ylabel("$ASK$")
ax[3].plot(t, ask)

plt.show()
```
