```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

plt.rcParams.update({"axes.grid": True, "font.size": 14})
```

### Slide 1

1. Gerar um sinal $s(t)$ composto pela somatória de 3 cossenos com amplitudes de $6V$, $4V$ e $2V$ e frequências de 1, 3, e 5 kHz, respectivamente.

2. Plotar em uma figura os três cossenos e o sinal $s(t)$ no domínio do tempo e da frequência.

```python
A1 = 6
A2 = 2
A3 = 4

f1 = int(1e3)
f2 = int(3e3)
f3 = int(5e3)

fs = int(100e3)
t = np.arange(0, 1, 1 / fs)
f = np.arange(-fs / 2, fs / 2, 1)

x1_t = A1 * np.cos(2 * np.pi * f1 * t)
X1_t = np.fft.fftshift(np.fft.fft(x1_t)) / len(x1_t)

x2_t = A2 * np.cos(2 * np.pi * f2 * t)
X2_t = np.fft.fftshift(np.fft.fft(x2_t)) / len(x2_t)

x3_t = A3 * np.cos(2 * np.pi * f3 * t)
X3_t = np.fft.fftshift(np.fft.fft(x3_t)) / len(x3_t)

s_t = x1_t + x2_t + x3_t
S_f = np.fft.fftshift(np.fft.fft(s_t)) / len(s_t)
```

```python
fig, axs = plt.subplots(4, 2, figsize=(18, 12))
fig.tight_layout(h_pad=3, w_pad=4)

axs[0, 0].plot(t, x1_t)
axs[0, 0].set_title("$x_1(t)$")
axs[0, 0].set_xlabel("Time [s]")
axs[0, 0].set_ylabel("Amplitude")
axs[0, 0].set_xlim(0, 5 * 1 / f1)

axs[0, 1].plot(f, np.abs(X1_t))
axs[0, 1].set_title("$X_1(f)$")
axs[0, 1].set_xlabel("Frequency [Hz]")
axs[0, 1].set_ylabel("Magnitude")
axs[0, 1].set_xlim(-10e3, 10e3)

axs[1, 0].plot(t, x2_t)
axs[1, 0].set_title("$x_2(t)$")
axs[1, 0].set_xlabel("Time [s]")
axs[1, 0].set_ylabel("Amplitude")
axs[1, 0].set_xlim(0, 5 * 1 / f1)

axs[1, 1].plot(f, np.abs(X2_t))
axs[1, 1].set_title("$X_2(f)$")
axs[1, 1].set_xlabel("Frequency [Hz]")
axs[1, 1].set_ylabel("Magnitude")
axs[1, 1].set_xlim(-10e3, 10e3)

axs[2, 0].plot(t, x3_t)
axs[2, 0].set_title("$x_3(t)$")
axs[2, 0].set_xlabel("Time [s]")
axs[2, 0].set_ylabel("Amplitude")
axs[2, 0].set_xlim(0, 5 * 1 / f1)

axs[2, 1].plot(f, np.abs(X3_t))
axs[2, 1].set_title("$X_3(f)$")
axs[2, 1].set_xlabel("Frequency [Hz]")
axs[2, 1].set_ylabel("Magnitude")
axs[2, 1].set_xlim(-10e3, 10e3)

axs[3, 0].plot(t, s_t)
axs[3, 0].set_title("$s(t)$")
axs[3, 0].set_xlabel("Time [s]")
axs[3, 0].set_ylabel("Amplitude")
axs[3, 0].set_xlim(0, 5 * 1 / f1)

axs[3, 1].plot(f, np.abs(S_f))
axs[3, 1].set_title("$S(f)$")
axs[3, 1].set_xlabel("Frequency [Hz]")
axs[3, 1].set_ylabel("Magnitude")
axs[3, 1].set_xlim(-10e3, 10e3)

plt.show()
```

3. Utilizando a função `norm`, determine a potência média do sinal $s(t)$.

```python
p_mean = np.linalg.norm(s_t) ** 2 / len(s_t)

p_mean
```

4. Utilizando a função `pwelch`, plote a densidade espectral de potência do sinal $s(t)$.

```python
f, Pxx = signal.welch(s_t, fs, nperseg=1024)
```

```python
plt.figure(figsize=(18, 4))
plt.plot(f, Pxx)

plt.title("Densidade Espectral de Potência")
plt.xlabel("Frequencia [Hz]")
plt.ylabel("Densidade Espectral de Potência [W/Hz]")
plt.xlim(0, 6e3)
plt.show()
```

### Slide 2

3. Gerar 3 filtros ideais:
    - Passa-baixa (frequência de corte de 2 kHz)
    - Passa-alta (frequência de corte de 4 kHz)
    - Passa-faixa (frequências de corte de 2 e 4 kHz)

4. Plotar em uma figura a resposta em frequência dos 3 filtros.

```python
fc_low = int(2e3)
fc_high = int(4e3)

LPF_f = np.concatenate(
    (
        np.zeros(int(fs / 2 - fc_low)),
        np.ones(fc_low * 2),
        np.zeros(int(fs / 2 - fc_low)),
    )
)
lpf_t = np.fft.ifft(np.fft.ifftshift(LPF_f))


HPF_f = np.concatenate(
    (
        np.ones(int(fs / 2 - fc_high)),
        np.zeros(fc_high * 2),
        np.ones(int(fs / 2 - fc_high)),
    )
)
hpf_t = np.fft.ifft(np.fft.ifftshift(HPF_f))

BPF_f = np.concatenate(
    (
        np.zeros(int(fs / 2 - fc_high)),
        np.ones(fc_high - fc_low),
        np.zeros(fc_low * 2),
        np.ones(fc_high - fc_low),
        np.zeros(int(fs / 2 - fc_high)),
    )
)
bpf_t = np.fft.ifft(np.fft.ifftshift(BPF_f))
```

```python
fig, axs = plt.subplots(3, 1, figsize=(18, 9))
fig.tight_layout(h_pad=3, w_pad=4)

axs[0].plot(f, np.abs(LPF_f))
axs[0].set_title("Filtro Passa-Baixa")
axs[0].set_ylabel("Magnitude")

axs[1].plot(f, np.abs(HPF_f))
axs[1].set_title("Filtro Passa-Alta")
axs[1].set_ylabel("Magnitude")

axs[2].plot(f, np.abs(BPF_f))
axs[2].set_title("Filtro Passa-Banda")
axs[2].set_xlabel("Frequência [Hz]")
axs[2].set_ylabel("Magnitude")

plt.show()
```

5. Passar o sinal $s(t)$ pelos 3 filtros e plotar as saídas, no domínio do tempo e da frequência, para os 3 casos.

```python
S_f_lpf = S_f * LPF_f
s_t_lpf = np.real(np.fft.ifft(np.fft.ifftshift(S_f_lpf)))

S_f_hpf = S_f * HPF_f
s_t_hpf = np.real(np.fft.ifft(np.fft.ifftshift(S_f_hpf)))

S_f_bpf = S_f * BPF_f
s_t_bpf = np.real(np.fft.ifft(np.fft.ifftshift(S_f_bpf)))
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 9))
fig.tight_layout(h_pad=3, w_pad=4)

axs[0, 0].plot(t, s_t_lpf)
axs[0, 0].set_title("Sinal Filtrado - Passa-Baixa")
axs[0, 0].set_ylabel("Amplitude")
axs[0, 0].set_xlim(0, 5 * 1 / f1)

axs[0, 1].plot(f, np.abs(S_f_lpf))
axs[0, 1].set_title("Espectro do Sinal Filtrado - Passa-Baixa")
axs[0, 1].set_ylabel("Magnitude")
axs[0, 1].set_xlim(-10e3, 10e3)

axs[1, 0].plot(t, s_t_hpf)
axs[1, 0].set_title("Sinal Filtrado - Passa-Alta")
axs[1, 0].set_ylabel("Amplitude")
axs[1, 0].set_xlim(0, 5 * 1 / f1)

axs[1, 1].plot(f, np.abs(S_f_hpf))
axs[1, 1].set_title("Espectro do Sinal Filtrado - Passa-Alta")
axs[1, 1].set_ylabel("Magnitude")
axs[1, 1].set_xlim(-10e3, 10e3)

axs[2, 0].plot(t, s_t_bpf)
axs[2, 0].set_title("Sinal Filtrado - Passa-Banda")
axs[2, 0].set_xlabel("Tempo [s]")
axs[2, 0].set_ylabel("Amplitude")
axs[2, 0].set_xlim(0, 5 * 1 / f1)

axs[2, 1].plot(f, np.abs(S_f_bpf))
axs[2, 1].set_title("Espectro do Sinal Filtrado - Passa-Banda")
axs[2, 1].set_xlabel("Frequência [Hz]")
axs[2, 1].set_ylabel("Magnitude")
axs[2, 1].set_xlim(-10e3, 10e3)

plt.show()
```

5. Utilizando a função `filtro = fir1(50, (1000*2)/fs)`, realize uma operação de filtragem passa baixa do ruído. Para visualizar a resposta em frequência do filtro projetado, utilize a função `freqz`.

```python
# 5. Utilizando a função `filtro = fir1(50, (1000*2)/fs)`, realize uma operação de filtragem passa baixa do ruído. Para visualizar a resposta em frequência do filtro projetado, utilize a função `freqz`.
ordem = 50
fc = 1000

b = signal.firwin(ordem + 1, (fc * 2) / fs)
w, h = signal.freqz(b)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(12, 6))
fig.tight_layout(h_pad=3, w_pad=4)

axs[0].plot(w, 20 * np.log10(abs(h)))
axs[0].set_title("Resposta em Frequência do Filtro")
axs[0].set_ylabel("Amplitude [dB]")
axs[0].set_xlabel("Frequência [rad/sample]")

axs[1].plot(w, np.unwrap(np.angle(h)))
axs[1].set_title("Fase do Filtro")
axs[1].set_ylabel("Fase (radianos)")
axs[1].set_xlabel("Frequência [rad/sample]")

plt.show()
```
