```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

from numpy.fft import fft, fftshift
```

```python
filename = "./data/gqrx.raw"
dataIQ = np.fromfile(filename, dtype=np.float32)
I = dataIQ[0::2]
Q = dataIQ[1::2]

C = I + 1j * Q

# Calcule a FFT do sinal complexo e aplique fftshift
fft_C = fftshift(fft(C))

fs = 1.8e6  # Freqüência de amostragem do sinal em Hz.
N = len(C)
frequencies = np.arange(-N / 2, N / 2) * fs / N

plt.plot(frequencies, abs(fft_C))
plt.xlabel("Frequência (Hz)")
plt.title("FFT do sinal")
plt.xlim(-150e3, 150e3)
plt.show()
```
